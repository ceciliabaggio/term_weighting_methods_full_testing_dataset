Funcionalidad para importar los datos de TEST de MOGP.
Las métricas poblacionales para el testing, como el Global Recall o Jaccard Index Sim, estaban en 1 solo archivo de 150 lineas (un valor por generación) por 5 columnas (corridas). Así que rescaté los valores de la ultima gen.
Las métricas individuales están para todos los individuos: 100 indiv*150 gen FILAS y 5 columnas.

```python
import os
import pandas as pd


## PRECISION @10

df_P_out = pd.DataFrame()

topic_number = []
for root, dirs, files in os.walk('MOGP/Co3/'):
     
    for file in files:
    
        if "MEAN_PRECISION@10_FROM_TESTING_GEN_LAST_array_of_mean" in file:            
            topic = root.split("/")[2]            
            topic_number.append(topic)
              
            df_precision_all_run = pd.read_csv(os.path.join(root, file), header=None, delim_whitespace=True, skip_blank_lines=True)                    
            df_precision_all_run = df_precision_all_run.transpose()
            df_precision_all_run['Topic_number'] = topic
            
            # # mover Topic ID a la primer col
            # get a list of columns
            cols = list(df_precision_all_run)
            # move the column to head of list using index, pop and insert
            cols.insert(0, cols.pop(cols.index('Topic_number')))
            #print(cols)

            # use ix to reorder
            df_precision_all_run = df_precision_all_run.loc[:, cols]
            df_P_out = df_P_out.append(pd.DataFrame(df_precision_all_run))

df_P_out.sort_values(by='Topic_number', axis=0, ascending=True, inplace=True, ignore_index = True)             
    
col_dict={}

col_dict[0] = "run_1"
col_dict[1] = "run_2"
col_dict[2] = "run_3"
col_dict[3] = "run_4"
col_dict[4] = "run_5"

df_P_out.rename(columns = col_dict, inplace = True)
df_P_out["mean"] = df_P_out.loc[:,[c for c in df_P_out.columns if c!= "Topic_number"]].mean(axis=1)
print(df_P_out)

df_P_out.to_csv('MOGP/Co3_MEAN_PRECISION@10_FROM_TESTING_ALL_TOPICS_LAST_GEN_ordered.csv', encoding='utf-8')
```

Funcionalidad para importar los datos de MOGP.
Las métricas poblacionales para el entrenamiento, como el Global Recall o Jaccard Index Sim, estaban en 1 solo archivo de 150 lineas (un valor por generación) por 5 columnas (corridas). Así que rescaté los valores de la ultima gen.
Las métricas individuales están para todos los individuos: 100 indiv*150 gen FILAS y 5 columnas.

```python
import os
import pandas as pd

# create an Empty DataFrame
# object With column names only
df_GR_out = pd.DataFrame(columns = ['Topic_number', 'GR_lastGen_run_1', 'GR_lastGen_run_2', 'GR_lastGen_run_3','GR_lastGen_run_4', 'GR_lastGen_run_5'])


for root, dirs, files in os.walk('MOGP/Co3/'):
     
    for file in files:
    
        if "globalRecall_all_run" in file:
            topic_number = file.split("_")[0]
            
            df_GR = pd.read_csv(os.path.join(root, file), header=None, delim_whitespace=True, skip_blank_lines=True)            
            last_row = df_GR.iloc[-1] # la ultima fila corresponde al GR de la ultima GEN del TRAINING
            #print(last_row[0], last_row[1], last_row[2], last_row[3], last_row[4] )
            
            
            # append rows to an empty DataFrame
            df_GR_out = df_GR_out.append({'Topic_number' : topic_number, 
                                          'GR_lastGen_run_1' : last_row[0], 
                                          'GR_lastGen_run_2' : last_row[1],
                                          'GR_lastGen_run_3' : last_row[2],
                                          'GR_lastGen_run_4' : last_row[3],
                                          'GR_lastGen_run_5' : last_row[4]}, 
                                        ignore_index = True)

df_GR_out.sort_values(by='Topic_number', axis=0, ascending=True, inplace=True, ignore_index = True)
print(df_GR_out)

#df_GR_out['mean'] = df_GR_out.mean(axis=1) # deberia expluir el topic ID
df_GR_out["mean"] = df_GR_out.loc[:,[c for c in df_GR_out.columns if c!= "Topic_number"]].mean(axis=1)
print(df_GR_out)
df_GR_out.to_csv('MOGP/Co3_globalRecall_allRun_lastGen_per_topic_TRAINING.csv', encoding='utf-8')


            
```

```python
import os
import pandas as pd


## PRECISION @10

df_P_out = pd.DataFrame()

topic_number = []
for root, dirs, files in os.walk('MOGP/Co3/'):
     
    for file in files:
    
        if "precision_all_run" in file:
            topic = file.split("_")[0]
            
            topic_number.append(topic)
            
            df_precision_all_run = pd.read_csv(os.path.join(root, file), header=None, delim_whitespace=True, skip_blank_lines=True)                    
            #df_precision_all_run.to_csv('MOGP/Co3_precision_PRUEBA'+ topic+'.csv', encoding='utf-8')
            last_100_rows = df_precision_all_run.iloc[-100:] # las ultimas 100 filas corresponde al GR de la ultima GEN del TRAINING        
            df_P_out = df_P_out.append(pd.DataFrame(last_100_rows.mean(axis=1)).transpose())

            
df_P_out['Topic_number'] = topic_number  
    
col_dict={}

for index, i in enumerate(range(14900,15000)):
    col_dict[i] = "query_"+ str(index) 

    df_P_out.rename(columns = col_dict, inplace = True)

df_P_out.sort_values(by='Topic_number', axis=0, ascending=True, inplace=True, ignore_index = True)            
#print(df_P_out)

# mover Topic ID a la primer col
# get a list of columns
cols = list(df_P_out)
# move the column to head of list using index, pop and insert
cols.insert(0, cols.pop(cols.index('Topic_number')))
#print(cols)

# use ix to reorder
df_P_out = df_P_out.loc[:, cols]

print(df_P_out)
df_P_out.to_csv('MOGP/Co3_precision_allRun_lastGen_per_topic_TRAINING.csv', encoding='utf-8')
```
